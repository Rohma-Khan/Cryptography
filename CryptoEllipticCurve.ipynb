```python
def egcd(a, n):
    """return (g, x, y) such that a*x + b*y = g = gcd(a, b)"""
    if a == 0:
        return (n, 0, 1)
    else:
        g, x, y = egcd(n % a, a)
        return (g, y - (n // a) * x, x)
def inverse(a,n):
    g,inv,garbage=egcd(a,n)
    if ((g%n)==1):
        return (inv%n)
    else:
        print ("the gcd of " ,a," and ",n, "is ", g)
        print ("Error, we have reached the point at infinity disregard everything bellow")
        return -1
```

```python
class EllipticCurve(object):
    def __init__(self, a,b,n):
        self.a=a
        self.b=b
        self.n=n
    def testPoint(self,x,y):
        return (y*y)%self.n==(x*x*x+(self.a)*x+(self.b))%self.n
    def singular(self):
        return 0==(4*pow((self.a),3)+27*pow((self.b),3))%self.n
    def add(self,x1,y1,x2,y2):
        if (x1==x2 and y1==y2):
            l=(((3*x1*x1+self.a)%self.n)*(inverse(2*y1,self.n)%self.n))%self.n
        else:
            l=(((y2-y1)%self.n)*(inverse((x2-x1)%self.n,self.n)%self.n))%self.n
        x3=((l*l)-x1-x2)%self.n
        y3=(l*(x1-x3)-y1)%self.n
        return x3,y3
    def NegativePoint (self,x,y):
        y=-y%self.n
        return x,y
```
